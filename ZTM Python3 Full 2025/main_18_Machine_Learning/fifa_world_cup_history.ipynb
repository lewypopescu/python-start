```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configure visualization style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
cups = pd.read_csv("WorldCups.csv")
matches = pd.read_csv("WorldCupMatches.csv")
players = pd.read_csv("WorldCupPlayers.csv")

print("✅ Data loaded successfully!")
print("WorldCups:", cups.shape)
print("Matches:", matches.shape)
print("Players:", players.shape)
```

```python
display(cups.head())
display(matches.head())
display(players.head())

```

```python
winners = cups['Winner'].value_counts()

plt.bar(winners.index, winners.values, color="skyblue")
plt.title("Most Successful World Cup Nations")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.show()

winners
```

```python
matches['TotalGoals'] = matches['Home Team Goals'] + matches['Away Team Goals']
goals_by_year = matches.groupby('Year')['TotalGoals'].sum()

plt.plot(goals_by_year.index, goals_by_year.values, marker='o', color='red')
plt.title("Total Goals per World Cup")
plt.xlabel("Year")
plt.ylabel("Goals Scored")
plt.show()

```

```python
plt.bar(cups['Country'], cups['Attendance'], color='lightgreen')
plt.xticks(rotation=90)
plt.title("Attendance by Host Country")
plt.ylabel("Total Attendance")
plt.show()
```

```python
matches_per_year = matches.groupby('Year')['MatchID'].count()
goals_per_match = goals_by_year / matches_per_year

plt.plot(goals_per_match.index, goals_per_match.values, marker='o', color='purple')
plt.title("Average Goals per Match Over Time")
plt.xlabel("Year")
plt.ylabel("Goals per Match")
plt.show()
```

## 📈 Insights Summary

- Brazil has won the most World Cups in history.  
- Total goals per tournament have slightly decreased after 1998.  
- Average goals per match are around 2.5–3.0 over the years.  
- Attendance peaked in tournaments hosted by the USA and Brazil.  
- Some legendary players (like Cafu, Klose, or Messi) appear multiple times.
