```python
def clean_currency(x):
    if isinstance(x, str):
        x = x.replace('€', '').replace('M', '000000').replace('K', '000')
        return float(x)
    return x

data_frame['Value_num'] = data_frame['Value'].apply(clean_currency)
data_frame['Wage_num'] = data_frame['Wage'].apply(clean_currency)
```

```python
import matplotlib.pyplot as plt

plt.scatter(data_frame['Value_num'], data_frame['Wage_num'], alpha=0.5)
plt.xlabel('Player Value (€)')
plt.ylabel('Player Wage (€)')
plt.title('Relation between Player Value and Wage')
plt.show()
```

```python
correlation = data_frame[['Value_num', 'Wage_num']].corr()
print(correlation)
```

```python
data_frame['Age'].hist(bins=20)
plt.title('Age Distribution of Players')
plt.xlabel('Age')
plt.ylabel('Number of Players')
plt.show()
```
